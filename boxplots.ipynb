```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import pie, axis, show


dataset = pd.read_csv('data/dataset.csv', 
    usecols=['Root Cause Category', 'Manifestation Category', 'Fix Category', 'How Fixed in Code', 'Testing Environment/Runner'], sep=',')
columns = dataset.columns.to_list()

def make_boxplots(dataset, column, title):

    count = dataset[column].value_counts().values
    categories = dataset[column].value_counts().index


    plt.figure(figsize = (15, 8))
    dataset[column].value_counts().plot(kind='box', title=column)
    plt.ylabel("Quantity")
    plt.legend()
```

```python
make_boxplots(dataset, 'Fix Category', 'Fix flaky methods')
```

```python
make_boxplots(dataset, 'Root Cause Category', 'Root Cause Category')
```

```python
make_boxplots(dataset, 'Manifestation Category', 'Manifestation Category')
```

```python
make_boxplots(dataset, 'Testing Environment/Runner', 'Testing Environment/Runner')
```
