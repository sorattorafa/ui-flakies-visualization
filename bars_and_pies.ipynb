```python

import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import pie, axis, show


dataset = pd.read_csv('dataset.csv', 
    usecols=['Root Cause Category', 'Manifestation Category', 'Fix Category', 'How Fixed in Code', 'Testing Environment/Runner'], sep=',')
columns = dataset.columns.to_list()

def make_quantity_plots(dataset, column, title):

    count = dataset[column].value_counts().values
    categories = dataset[column].value_counts().index


    plt.figure(figsize = (30, 8))
    plt.subplot(1, 2, 1)
    dataset[column].value_counts().plot(kind='bar', title=column)
    plt.xlabel(column)
    plt.ylabel("Quantity")
    plt.title(title)
    plt.subplot(1, 2, 2)
    plt.pie(count, 
            labels = list(categories),  
            colors = ["#20257c", "#424ad1", "#6a8ee8", "#66bbe2", "#66dee2", "#6ce2cb", "#6ad187", "#3b7f5b"],
            labeldistance = 1.8,
            wedgeprops = {"ec": "k"}, 
            textprops = {"fontsize": 11}, 
            )
    plt.axis("equal")
    plt.title(title)
    plt.legend()
```

```python
make_quantity_plots(dataset, 'Fix Category', 'Fix flaky methods')
```

```python

make_quantity_plots(dataset, 'Root Cause Category', 'Root Cause Category')
```

```python
make_quantity_plots(dataset, 'Manifestation Category', 'Manifestation Category')
```

```python
make_quantity_plots(dataset, 'Testing Environment/Runner', 'Testing Environment/Runner')
```
